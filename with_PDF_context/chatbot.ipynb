# Winzerfest Chatbot

```python
from dotenv import load_dotenv
from openai import OpenAI
from pypdf import PdfReader
import gradio as gr
```

```python
import os
load_dotenv(override=True)
google_api_key = os.getenv("GEMINI_API_KEY")
GEMINI_BASE_URL = os.getenv("GEMINI_BASE_URL")
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
```

```python
reader = PdfReader("data/programm.pdf")
program = ""

for page in reader.pages: 
        text = page.extract_text()
        if text:
            program += text
```

```python
system_prompt = f"You are acting as a local guide for the Besigheimer Winzerfest. You are answering questions on the programm of the festival. \
    You are given a overview of the program and the offerings of the festival which you can use to answer questions. \
    Be friendly and engaging. If you don't know the answer, say so."

system_prompt += f"\n\n Program:\n {program}"
system_prompt += f"\n\n With this context, please chat with the user."
```

```python
system_prompt
```

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]
    response = gemini.chat.completions.create(
        model="gemini-2.5-flash-preview-05-20",
        messages=messages,
        temperature=0.7,
    )
    return response.choices[0].message.content
```

```python
gr.ChatInterface(chat, type="messages").launch()
```
